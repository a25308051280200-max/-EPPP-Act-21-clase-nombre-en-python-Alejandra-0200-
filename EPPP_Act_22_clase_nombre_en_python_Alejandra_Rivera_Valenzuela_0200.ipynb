```python
# Definición de la Clase
class Deportista:
    def __init__(self, nombre, compromiso, inteligencia):
        self.nombre = nombre
        self.compromiso = compromiso
        self.inteligencia = inteligencia

    # --- Métodos (Acciones) ---
    def correr(self):
        print(f"\n🏃 {self.nombre} está corriendo en la pista de atletismo.")

    def calentar(self):
        print(f"\n🔥 {self.nombre} está calentando. Inteligencia aplicada: {self.inteligencia}.")

    def entrenamientos(self):
        print(f"\n💪 {self.nombre} inició entrenamiento. Compromiso: {self.compromiso}.")

# --- Función Principal ---
def sistema_deportivo():
    deportistas = [] # Lista para guardar a las dos deportistas

    # Registro de las 2 deportistas
    for i in range(1, 3):
        print(f"\n--- REGISTRO DEPORTISTA {i} ---")
        nom = input(f"Nombre de la deportista {i}: ")
        comp = input(f"Nivel de compromiso de {nom}: ")
        intel = input(f"Tipo de inteligencia de {nom}: ")

        # Creamos el objeto y lo guardamos en la lista
        nuevo_deportista = Deportista(nom, comp, intel)
        deportistas.append(nuevo_deportista)

    # Menú Interactivo
    while True:
        print("\n" + "═"*35)
        print("       MENÚ DE SELECCIÓN")
        print("" + "═"*35)
        print(f"1. Gestionar a {deportistas[0].nombre}")
        print(f"2. Gestionar a {deportistas[1].nombre}")
        print("3. Salir")

        opc_persona = input("\n¿A quién quieres ver? (1-3): ")

        if opc_persona in ["1", "2"]:
            # Elegimos a la deportista según el índice (0 o 1)
            d = deportistas[int(opc_persona) - 1]

            print(f"\n--- Acciones para {d.nombre} ---")
            print("a) Calentar")
            print("b) Correr")
            print("c) Entrenar")
            accion = input("Selecciona acción (a, b, c): ").lower()

            if accion == "a": d.calentar()
            elif accion == "b": d.correr()
            elif accion == "c": d.entrenamientos()
            else: print("Acción no válida.")

        elif opc_persona == "3":
            print("\nCerrando sistema deportivo...")
            break
        else:
            print("\n[!] Opción no válida.")

# --- Ejecución ---
sistema_deportivo()

print("\n" + "═"*40)
print("Programa terminado con éxito")
print("Creador de software por Alejandra Rivera Valenzuela NL = 45")
print("═"*40)
```
